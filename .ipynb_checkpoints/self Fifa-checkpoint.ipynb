```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fifa = pd.read_csv('fifa data.csv')
```

```python
fifa.head()
```

```python
for col in fifa:
    print(col)
```

```python
fifa['nationality'].value_counts()
```

```python
fifa['nationality'].value_counts()[0:10]

```

```python
fifa['nationality'].value_counts()[0:5]
    
```

```python
plt.bar(fifa['nationality'].value_counts()[0:5].keys(), (fifa['nationality'].value_counts()[0:5]), color = 'purple')
plt.title('Nationality of Team')
plt.xlabel('Team')
plt.ylabel('Count')
plt.show()
```

```python
fifa.head()
```

```python
player_sallery = fifa[['short_name', 'wage_eur']]
```

```python
player_sallery.head()
```

```python
player_sallery = player_sallery.sort_values(['wage_eur'], ascending=False)
```

```python
player_sallery.head()
```

```python
plt.bar((player_sallery['short_name'])[0:5],(player_sallery['wage_eur'])[0:5], color='red')
plt.title('Sallery Of Player')
plt.xlabel('Players')
plt.ylabel('Sallery')
plt.show()
```
